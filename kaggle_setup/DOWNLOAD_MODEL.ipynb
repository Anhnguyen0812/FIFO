# Download FIFO Trained Model

This notebook copies the trained model to /kaggle/working for download.

```python
import os
import shutil

# Source file path
source = '/kaggle/working/snapshots/FIFO_model/fast_training10000.pth'

# Destination (must be in /kaggle/working root)
dest = '/kaggle/working/fast_training10000.pth'

print(f"Copying model...")
print(f"From: {source}")
print(f"To: {dest}")

# Copy file
if os.path.exists(source):
    shutil.copy2(source, dest)
    print(f"‚úÖ Copy successful!")
    
    # Show file info
    size_mb = os.path.getsize(dest) / (1024*1024)
    print(f"üìä File size: {size_mb:.2f} MB")
    print(f"üìÅ Location: {dest}")
else:
    print(f"‚ùå Source file not found: {source}")
    print("\nAvailable files in snapshots:")
    !ls -lh /kaggle/working/snapshots/FIFO_model/ 2>/dev/null || echo "Snapshots folder not found"
```
