## üìã Cell 11: Check Results

```python
import glob
import os

snapshot_dir = '/kaggle/working/snapshots/FIFO_test'
checkpoints = sorted(glob.glob(f'{snapshot_dir}/*.pth'))

print(f"Found {len(checkpoints)} checkpoint(s):\n")

for ckpt in checkpoints:
    name = os.path.basename(ckpt)
    size = os.path.getsize(ckpt) / (1024**2)  # MB
    print(f"  ‚úì {name}")
    print(f"    Size: {size:.2f} MB\n")

if checkpoints:
    print("\n‚úÖ TEST SUCCESSFUL!")
    print("\nNext step: Create new notebook for FULL TRAINING")
    print("See: kaggle_setup/KAGGLE_NOTEBOOK_SETUP.md")
else:
    print("\n‚ö†Ô∏è No checkpoints found!")
    print("Please check the training output above for errors.")
```

## üìã Cell 12: Validate Checkpoint

```python
import torch
import glob

checkpoints = sorted(glob.glob('/kaggle/working/snapshots/FIFO_test/*.pth'))

if checkpoints:
    latest_ckpt = checkpoints[-1]
    print(f"Loading checkpoint: {os.path.basename(latest_ckpt)}")
    
    try:
        checkpoint = torch.load(latest_ckpt, map_location='cpu')
        print(f"\nCheckpoint keys: {list(checkpoint.keys())}")
        print(f"Training iteration: {checkpoint.get('train_iter', 'N/A')}")
        print("\n‚úì Checkpoint is valid!")
    except Exception as e:
        print(f"\n‚úó Error loading checkpoint: {e}")
else:
    print("No checkpoints found to validate")
```
